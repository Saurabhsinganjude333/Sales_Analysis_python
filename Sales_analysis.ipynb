```python
combined_df = pd.read_csv(r'C:\Users\sss923875\Downloads\python_projects\Sales_analysis\combined_data\all_data_copy.csv')
all_data = combined_df.dropna()
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

### change data type of column 

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'], format='%m/%d/%y %H:%M')
all_data.dtypes
```

### ADD NEW COLUMN  (MONTH COLUMN)

```python
all_data['Month'] = all_data['Order Date'].dt.month
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Add city column 

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1].strip())
all_data.head()
```

### Add sales column

```python
all_data['Sale'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
monthly_sales = all_data.groupby('Month')['Sale'].sum()
top_6_months = monthly_sales.sort_values(ascending=False).head(6)

# Plot the top 6 months
plt.figure(figsize=(10, 6))
ax = top_6_months.plot(kind='bar', color='skyblue')
plt.title('Top 6 Months for Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.grid(axis='y')

for i, value in enumerate(top_6_months):
    ax.text(i, value, f'${value:.2f}', ha='center', va='bottom')

plt.show()
```

```python
monthly_sales = all_data.groupby('Month')['Sale'].sum()
monthly_sales = monthly_sales.sort_index()
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(monthly_sales.index, monthly_sales, color='darkblue', marker='o', linestyle='-', label='Monthly Sales')
ax1.set_xlabel('Month')
ax1.set_ylabel('Total Sales ($)')
ax1.set_title('Monthly Sales')
ax1.grid(True)
ax2 = ax1.twinx()

bars = ax2.bar(monthly_sales.index, monthly_sales, color='skyblue', alpha=0.7, width=0.4, label='Sales')
ax2.set_ylabel('Total Sales ($)')
ax2.set_ylim(0, monthly_sales.max() * 1.1)


for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width() / 2, height, '', ha='center', va='bottom')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')
plt.show()
```

  <font color='blue' size=4> <b>Question 4: What products are most often sold together? </b></font>

```python
from itertools import combinations
from collections import Counter


grouped = all_data.groupby('Order ID')['Product'].apply(list)
pair_counts = Counter()

for order_products in grouped:
    pair_counts.update(combinations(order_products, 2))

most_common_pairs = pair_counts.most_common(10)  

for pair, count in most_common_pairs:
    print(pair, count)
```

```python
grouped = all_data.groupby('Order ID')['Product'].apply(list)
pair_counts = Counter()

for order_products in grouped:
 pair_counts.update(combinations(order_products, 2))
 most_common_pairs = pair_counts.most_common(15)  
products, counts = zip(*most_common_pairs)
product_pairs_str = [f"{pair[0]} + {pair[1]}" for pair in products]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(product_pairs_str, counts, color='skyblue')
plt.title('Most Often Sold Together Products')
plt.xlabel('Product Pairs')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

  <font color='blue' size=4> <b>Question 5: What product sold the most? Why do you think it sold the most? </b></font>

```python

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'], errors='coerce')

# Group by Product
product_group = all_data.groupby('Product')
quantity_ordered = product_group['Quantity Ordered'].sum()

quantity_ordered_sorted = quantity_ordered.sort_values(ascending=False)

# Get the sorted keys (product names)
keys = list(quantity_ordered_sorted.index)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(keys, quantity_ordered_sorted, color='skyblue')
plt.xticks(keys, rotation='vertical', size=8)
plt.xlabel('Product')
plt.ylabel('Total Quantity Ordered')
plt.title(' products with the highest total sales volume or revenue.')
plt.tight_layout()
plt.show()
```
